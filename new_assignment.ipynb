```
pivot_df = spark.read.csv("dbfs:/FileStore/tables/Pyspark_Task-1.csv",header=True,inferSchema = True)
pivot_df.show()
```

```
from pyspark.sql import *
from pyspark.sql import SparkSession
from pyspark.sql.functions import expr
```

```
unpivotExpr = "stack(5,'PL1',PL1,'PL2',PL2,'PL3',PL3,'PL4',PL4,'PL5',PL5) as (PlantName,Total)"
unPivotDF = pivot_df.select("Product", expr(unpivotExpr)) \
    .where("Total is not null")
unPivotDF.show(truncate=False)
unPivotDF.show()
```
