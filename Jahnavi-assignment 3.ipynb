```
# a) Create HR_Employee Pysaprk dataFrame. 
employee_DF = spark.read.csv("dbfs:/FileStore/tables/HR_Employee.csv",header=True,inferSchema = True)
employee_DF.show()
```

```
# b) Show printSchema & show 10 rows of Dataset
employee_DF.printSchema()
employee_DF.show(10, truncate=False)
```

```
# c) Most employee is working in which department
from pyspark.sql.functions import col
employee_DF.createOrReplaceTempView("employeetable")
spark.sql("select DEPARTMENT, COUNT(EmployeeID) as total_employees from employeetable group by DEPARTMENT order by total_employees desc").show(1)
```

```
# d) Highest number of job roles
spark.sql("select JobRole, COUNT(JobRole) as total_job_Roles from employeetable group by JobRole order by total_job_Roles desc").show()
```

```
# e) Which gender have higher strength as workforce?
spark.sql("select Gender, COUNT(Gender) as count_gender from employeetable group by Gender order by count_gender desc").show()
```

```
# f) Compare the marital status of employee and find the most frequent status.
#spark.sql("select MaritalStatus, COUNT(MaritalStatus) as Frequency from employeetable group by MaritalStatus order by count_status desc").show()
spark.sql("select MaritalStatus, COUNT(*) as Frequency from employeetable group by MaritalStatus order by count(*) desc limit 1").show()
```

```
# g) Mostly hired employee have qualification
spark.sql("select Education, COUNT(*) as Qualification from employeetable group by Education order by count(*) desc limit 1").show()
spark.sql("select Education, COUNT(Education) as Qualification from employeetable group by Education order by Qualification desc").show()
```

```
# h) Find the count of employee from which education fields 
spark.sql("select Education, COUNT(EmployeeID) as no_of_emp from employeetable group by Education order by no_of_emp desc").show()
```

```
# i) What is the job satisfaction level of employee? 
spark.sql("select JobSatisfaction, COUNT(JobSatisfaction) as no_of_employees from employeetable group by JobSatisfaction order by no_of_employees desc").show()
```

```
# j) Does most of employee do overtime: Yes or No?
spark.sql("select OverTime, COUNT(*) as no_of_employees from employeetable group by OverTime order by count(*) desc limit 1").show()
```

```
# k) Find Min & Max Salaried employees. 
employee_DF.agg({'Income' : 'max'}).show(1)
employee_DF.agg({'Income' : 'min'}).show(1)
```

```
# l) Does most of the employee do business travel? Find of the employees counts for each category
spark.sql("select BusinessTravel, COUNT(BusinessTravel) as no_of_employees from employeetable group by BusinessTravel order by no_of_employees desc").show()
spark.sql("select BusinessTravel, COUNT(*) as no_of_employees from employeetable group by BusinessTravel order by count(*) desc limit 1").show()
```

```
# m) Find the AVG salary of graduate employee.
spark.sql("select Education, AVG(Income) as average_salary from employeetable where Education='Bachelor' group by Education").show()
```

```
# n) Find the employee qualification receiving salary lower than equal to avg. salary of all employee.
avg_sal=spark.sql("select AVG(Income) from employeetable").show()
spark.sql("select Education, AVG(Income) as avg_salary from employeetable where Income <= (select AVG(Income) from employeetable) group by Education").show()
```

```
# p) Highest attrition is in which department? 
spark.sql("select Department, max(Attrition) as highest_attrition from employeetable group by Department").show()
```
